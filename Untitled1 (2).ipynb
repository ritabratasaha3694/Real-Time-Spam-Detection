```python
from nltk.corpus import wordnet
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
def get_wordnet_pos(treebank_tag):

    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

def process_text(text):

    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))
    #Remove special characters and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    #Convert to lowercase
    text = text.lower()
    #Tokenize
    tokens = word_tokenize(text)
    #POS tagging
    pos_tags = pos_tag(tokens)

    lemmatized_tokens = [
        lemmatizer.lemmatize(token, get_wordnet_pos(pos_tag))
        for token, pos_tag in pos_tags if token not in stop_words
    ]

    processed_text = ' '.join(lemmatized_tokens)
    return processed_text
```

```python
# Process text data for training set
X_train = X_train_raw.apply(process_text)
```

**TRAIN MODEL ON TRAINING SET**

```python
from sklearn.metrics import accuracy_score, classification_report, precision_score, recall_score, f1_score
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform labels
y_train= label_encoder.fit_transform(y_train)
```

```python
#Training naive bayes model
nb_model = MultinomialNB()
nb_model.fit(X_train, y_train)
```

```python
#Training svm model
svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)
```

```python
#Training random forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
```

```python
#Training logistic regression
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train, y_train)
```

**TESTING TEXT DATA FROM TESTSET**

```python
X_test = X_test_raw.apply(process_text)
X_test = vectorizer.transform(X_test)
```

```python
# Predictions
svm_pred = svm_model.predict(X_test)
nb_pred = nb_model.predict(X_test)
rf_pred = rf_model.predict(X_test)
lr_pred = lr_model.predict(X_test)
```

```python
label_encoder = LabelEncoder()
# Fit and transform labels
y_test= label_encoder.fit_transform(y_test)
```

```python
y_test
```

```python
# Define a function to evaluate each model and print metrics
def evaluate_model(model, X_test, y_test):
    # Predictions
    y_pred = model.predict(X_test)

    # Accuracy
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {accuracy}")

    # Precision
    precision = precision_score(y_test, y_pred, average='weighted')
    print(f"Precision: {precision}")

    # Recall
    recall = recall_score(y_test, y_pred, average='weighted')
    print(f"Recall: {recall}")

    # F1 Score
    f1 = f1_score(y_test, y_pred, average='weighted')
    print(f"F1 Score: {f1}")

    # Classification Report
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
```

```python
# Evaluate SVM model
print("Support Vector Machine (SVM):")
evaluate_model(svm_model, X_test, y_test)

# Evaluate Naive Bayes model
print("Naive Bayes:")
evaluate_model(nb_model, X_test, y_test)

# Evaluate Random Forest model
print("Random Forest:")
evaluate_model(rf_model, X_test, y_test)

# Evaluate Logistic Regression model
print("Logistic Regression:")
evaluate_model(lr_model, X_test, y_test)
```
